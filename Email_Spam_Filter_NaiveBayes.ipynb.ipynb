```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

df['text'] = df['text'].fillna('')
df.dropna(subset=['text'], inplace=True)

x = df.text.values
y = df.spam.values

xtrain, xtext, ytrain, ytext = train_test_split(x, y, test_size=0.2, random_state=42)

cv = CountVectorizer()
x = cv.fit_transform(xtrain)

model = MultinomialNB()
model.fit(x, ytrain)
```

```python
x_test = cv.transform(xtext)
y_pred = model.predict(x_test)
accuracy = accuracy_score(ytext, y_pred)
print("Accuracy:", accuracy)
```

```python
test_email = ["Subject: Hello My Friend"]
test_vector = cv.transform(test_email)
result = model.predict(test_vector)

print("\nThe test email is:", "Spam" if result[0] == 1 else "Not Spam")
```
